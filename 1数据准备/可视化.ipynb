```python
years , frequency = zip(*sorted(numPapersByYearCounter.items()))
fig, ax = plt.subplots()
ax.plot(years,np.array(frequency)/10000 )
# ax.set(xlabel='时间（年）', ylabel='总发文量',
#        title='期刊发文量随时间变化图', font=prop.get_name())
ax.set_ylabel('总发文量', font=prop.get_name())
ax.set_xlabel('时间（年）', font=prop.get_name())
ax.set_title('期刊发文量随时间变化图', font=prop.get_name())
plt.ticklabel_format(style='plain')
plt.text(-0.1,1.05,'(A)', color = 'red',
horizontalalignment='center',verticalalignment='center',
transform=ax.transAxes)
plt.text(.05,1.07,'(万篇)',
horizontalalignment='center',verticalalignment='center',
transform=ax.transAxes)
ax.set_xlim(1950, 2022)
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
fig.savefig("期刊发文量随时间变化图.png",dpi=300,bbox_inches='tight')
plt.show()
```
