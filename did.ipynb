```r
  library(did)
```

```r
library(rempsyc) #导出表格用
```

```r
mngAuthor = read.csv('did_new_2005_2015.csv')
```

```r
mngAuthor
```

```r
# 对生产力的影响
```

```r
outProductivity <- att_gt(yname = "avgPubNumAsofYear",
              gname = "firstTreat",
              idname = "authorID",
              tname = "year",
              xformla =  ~avg5yearCitationNumAsofYear+avgCoauthorNumAsofYear+creerLength+avgIDRpercentileAsofYear+avgRefsAsofYear ,
              data = mngAuthor,
              est_method = "dr",
              )
```

```r
summary(outProductivity)
```

```r
es <- aggte(outProductivity, type = "dynamic")
```

```r
summary(es)
```

```r
ggdid(es)
```

```r
p <- ggdid(es)
png("mtcars.png", width = 1200, height = 1200, res=200)
print(p)
dev.off()
```

```r
group_effects <- aggte(outProductivity, type = "group")
summary(group_effects)
```
