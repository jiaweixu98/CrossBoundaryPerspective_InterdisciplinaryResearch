```python
import numpy as np
# plot 同时中英混排
import matplotlib.pyplot as plt
from matplotlib import font_manager
import pickle as pk
plt.style.use('seaborn')
plt.figure(figsize=(4,3), dpi=300)
# 字体加载
font_path = "../../font/tnw+simsun.ttf"
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)
# 字体设置
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = prop.get_name()  # 根据名称设置字体
plt.rcParams['font.size'] = 12 # 设置字体大小
plt.rcParams['axes.unicode_minus'] = False # 使坐标轴刻度标签正常显示正负号
```

```python
numPapersByYearCounter = pk.load(open('../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/numPapersByYearCounter.pk', 'rb'))

numAvgreferencesCounter = pk.load( open('../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/numAvgreferencesCounter.pk', 'rb'))

numAvgcitationCountCounter = pk.load( open('../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/numAvgcitationCountCounter.pk', 'rb'))

numJournalCountCounter=  pk.load(open('../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/numJournalCountCounter.pk','rb'))

numCategoryCounter = pk.load(open('../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/numCategoryCounter.pk','rb'))
```

```python
numPapersByYearCounter_new = {}
for k,v in numPapersByYearCounter.items():
    if  1800 < k < 2022:
        numPapersByYearCounter_new[k] = v
        numPapersByYearCounter = numPapersByYearCounter_new
numPapersByYearCounter_new = {}
for k,v in numJournalCountCounter.items():
    if  1800 < k < 2022:
        numPapersByYearCounter_new[k] = v
        numJournalCountCounter = numPapersByYearCounter_new

numPapersByYearCounter_new = {}
for k,v in numCategoryCounter.items():
    if  1800 < k < 2022:
        numPapersByYearCounter_new[k] = v
        numCategoryCounter = numPapersByYearCounter_new

numPapersByYearCounter_new = {}
for k,v in numAvgreferencesCounter.items():
    if  1800 < k < 2022:
        numPapersByYearCounter_new[k] = v
        numAvgreferencesCounter = numPapersByYearCounter_new

numPapersByYearCounter_new = {}
for k,v in numAvgcitationCountCounter.items():
    if  1800 < k < 2022:
        numPapersByYearCounter_new[k] = v
        numAvgcitationCountCounter = numPapersByYearCounter_new
```

```python
years , frequency = zip(*sorted(numPapersByYearCounter.items()))
fig, ax = plt.subplots()
ax.plot(years,np.array(frequency)/10000 )
ax.set(xlabel='时间（年）', ylabel='总发文量',
       title='期刊发文量随时间变化图')
plt.ticklabel_format(style='plain')

plt.text(-0.03,1.07,'(万篇)',
horizontalalignment='center',verticalalignment='center',
transform=ax.transAxes)
ax.set_xlim(1950, 2022)
fig.savefig("期刊发文量随时间变化图.png",dpi=300)
plt.show()
```

```python
years , JournalCount = zip(*sorted(numJournalCountCounter.items()))
fig, ax = plt.subplots()
ax.plot(years,JournalCount)
ax.set(xlabel='时间（年）', ylabel='JCR期刊数量',
       title='JCR期刊数量随时间变化图')
plt.ticklabel_format(style='plain')

# plt.text(-0.03,1.07,'(篇)',
# horizontalalignment='center',verticalalignment='center',
# transform=ax.transAxes)
ax.set_xlim(1950, 2022)
fig.savefig("JCR期刊数量随时间变化图.png",dpi=300)
plt.show()
```

```python
years , categoryCount = zip(*sorted(numCategoryCounter.items()))
fig, ax = plt.subplots()
ax.plot(years,categoryCount)
ax.set(xlabel='时间（年）', ylabel='JCR学科数量',
       title='JCR学科数量随时间变化图')
plt.ticklabel_format(style='plain')

# plt.text(-0.03,1.07,'(篇)',
# horizontalalignment='center',verticalalignment='center',
# transform=ax.transAxes)
ax.set_xlim(1800, 2022)
fig.savefig("学科数量随时间变化图.png",dpi=300)
plt.show()
```

```python
years , citationcount = zip(*sorted(numAvgcitationCountCounter.items()))
years , refcount = zip(*sorted(numAvgreferencesCounter.items()))

fig, ax = plt.subplots()
ax.plot(years,citationcount, label='平均被引量数量')
ax.plot(years,refcount, linestyle='dashed', label='平均参考文献数量')
ax.set(xlabel='时间（年）', ylabel='平均被引量或参考文献数量（篇）',
       title='平均被引量学科数量随时间变化图')
plt.ticklabel_format(style='plain')

# plt.text(-0.03,1.07,'(篇)',
# horizontalalignment='center',verticalalignment='center',
# transform=ax.transAxes)
ax.set_xlim(1950, 2022)
ax.legend()
fig.savefig("平均被引量或参考文献数量随时间变化图.png",dpi=300)
plt.show()
```
