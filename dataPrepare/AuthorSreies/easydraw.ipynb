```python
import numpy as np
# plot 同时中英混排
import matplotlib.pyplot as plt
from matplotlib import font_manager
import matplotlib.ticker as mtick
import pickle as pk
plt.style.use('seaborn')
plt.figure(figsize=(4,3), dpi=300)
# 字体加载
font_path = "../../../font/tnw+simsun.ttf"
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)
# 字体设置
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = prop.get_name()  # 根据名称设置字体
plt.rcParams['font.size'] = 12 # 设置字体大小
plt.rcParams['axes.unicode_minus'] = False # 使坐标轴刻度标签正常显示正负号
```

```python
authornumPubYearsFreq = pk.load( open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/authornumPubYearsFreq.pk', 'rb'))
Rank1authornumPubYearsFreq = pk.load(open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/Rank1authornumPubYearsFreq.pk', 'rb'))
RankLasTauthornumPubYearsFreq = pk.load(open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/RankLasTauthornumPubYearsFreq.pk', 'rb'))
RankIMportantAuthornumPubYearsFreq = pk.load(open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/RankIMportantAuthornumPubYearsFreq.pk', 'rb'))
```

```python
authornumPubYearsFreq[0] = 0
freq1 , allalthor = zip(*sorted(authornumPubYearsFreq.items()))
freq2 , rank1 = zip(*sorted(Rank1authornumPubYearsFreq.items()))
freq3 , ranklast = zip(*sorted(RankLasTauthornumPubYearsFreq.items()))
freq4 , rankImportant = zip(*sorted(RankIMportantAuthornumPubYearsFreq.items()))

fig, ax = plt.subplots()
ax.plot(freq1,np.array(allalthor)/56523870,  linestyle='dashed', label='任意位置署名作者',marker='.',markersize=14)
ax.plot(freq4,np.array(rankImportant)/56523870,linestyle='dashed', label='末位作者或第一作者',marker='.',markersize=14)
ax.plot(freq2,np.array(rank1)/56523870, label='第一作者',marker='.',markersize=14)
ax.plot(freq3,np.array(ranklast)/56523870,  label='末位作者',marker='.',markersize=14)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
ax.set(xlabel='作者以某身份发文的总年数(年)', ylabel='比例',
       title='作者发文总年数分布: 参与者、第一作者、末位作者身份 (0至3年)')
# plt.ticklabel_format(style='plain')

plt.text(-0.03,1.05,'总人数:5652万',
horizontalalignment='center',verticalalignment='center',
transform=ax.transAxes)
ax.set_xlim(- 0.1, 3.1)
ax.set_xticks([0,1,2,3])
# ax.set_ylim(0, 4400)
# ax.set_yscale('log')
ax.legend(fontsize=12,framealpha=0.7,facecolor='white',frameon=True)
fig.savefig("作者发文总年数分布-参与者、第一作者、末位作者身份 (0至3年).png",dpi=300)
plt.show()
```
