```python
import numpy as np
# plot 同时中英混排
import matplotlib.pyplot as plt
from matplotlib import font_manager
import matplotlib.ticker as mtick
import pickle as pk
plt.style.use('seaborn')
plt.figure(figsize=(4,3), dpi=300)
# 字体加载
font_path = "../../../font/tnw+simsun.ttf"
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)
# 字体设置
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = prop.get_name()  # 根据名称设置字体
plt.rcParams['font.size'] = 12 # 设置字体大小
plt.rcParams['axes.unicode_minus'] = False # 使坐标轴刻度标签正常显示正负号
```

```python
#计算中科院分区后的情况
authornumPubYearsFreq = pk.load( open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASauthornumPubYearsFreq.pk', 'rb'))
Rank1authornumPubYearsFreq = pk.load(open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASRank1authornumPubYearsFreq.pk', 'rb'))
RankLasTauthornumPubYearsFreq = pk.load(open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASRankLasTauthornumPubYearsFreq.pk', 'rb'))
RankIMportantAuthornumPubYearsFreq = pk.load(open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASRankIMportantAuthornumPubYearsFreq.pk', 'rb'))
RankNotIMportantAuthornumPubYearsFreq= pk.load( open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASRankNotIMportantAuthornumPubYearsFreq.pk', 'rb'))
```

```python
authornumPubYearsFreq[0] = 0
freq1 , allalthor = zip(*sorted(authornumPubYearsFreq.items()))
freq2 , rank1 = zip(*sorted(Rank1authornumPubYearsFreq.items()))
freq3 , ranklast = zip(*sorted(RankLasTauthornumPubYearsFreq.items()))
freq4 , rankImportant = zip(*sorted(RankIMportantAuthornumPubYearsFreq.items()))
freq5 , rankNotImportant = zip(*sorted(RankNotIMportantAuthornumPubYearsFreq.items()))


fig, ax = plt.subplots()
ax.plot(freq3,np.cumsum(np.array(ranklast))/35134287,  label='末位作者',marker='.',markersize=14)
ax.plot(freq2,np.cumsum(np.array(rank1))/35134287, label='第一作者',marker='.',markersize=14)
ax.plot(freq4,np.cumsum(np.array(rankImportant))/35134287,linestyle='dashed', label='末位作者或第一作者',marker='.',markersize=14)
ax.plot(freq5,np.cumsum(np.array(rankNotImportant))/35134287,  label='非第一作者或末位作者',marker='.',markersize=14)
ax.plot(freq1,np.cumsum(np.array(allalthor))/35134287,  linestyle='dashed', label='任意位置署名作者',marker='.',markersize=14)

ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
ax.set(xlabel='作者以某身份发文的总年数(年)', ylabel='累计比例',
       title='作者发文总年数分布: 参与者、第一作者、末位作者身份 (0至3年)')
# plt.ticklabel_format(style='plain')

plt.text(-0.03,1.05,'总人数:3513万',
horizontalalignment='center',verticalalignment='center',
transform=ax.transAxes)
ax.set_xlim(- 0.1, 3.1)
ax.set_xticks([0,1,2,3])
ax.set_ylim(-0.01, 1.01)
# ax.set_yscale('log')
ax.legend(fontsize=12,framealpha=0.7,facecolor='white',frameon=True)
fig.savefig("作者发文总年数分布-参与者、第一作者、末位作者身份 (0至3年).png",dpi=300)
plt.show()
```

```python
CommonStartYearCounter=pk.load( open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASCommonStartYearCounter.pk', 'rb'))
CommonEndYearCounter=pk.load( open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASCommonEndYearCounter.pk', 'rb'))
CommonSpanYearCounter=pk.load( open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASCommonSpanYearCounter.pk', 'rb'))
KeyStartYearCounter=pk.load( open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASKeyStartYearCounter.pk', 'rb'))
KeyEndYearCounter=pk.load( open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASKeyEndYearCounter.pk', 'rb'))
KeySpanYearCounter=pk.load( open('../../../DataCrossBoundaryPerspective_InterdisciplinaryResearch/CASKeySpanYearCounter.pk', 'rb'))
```

```python
# 开始或终止年份
CommonStartYear , freq1 = zip(*sorted(CommonStartYearCounter.items()))
CommonEndYear , freq2 = zip(*sorted(CommonEndYearCounter.items()))
KeyStartYear , freq3 = zip(*sorted(KeyStartYearCounter.items()))
KeyEndYear , freq4 = zip(*sorted(KeyEndYearCounter.items()))

fig, ax = plt.subplots()
ax.plot(CommonStartYear,np.cumsum(np.array(freq1))/5866212.0,  linestyle='dashed', label='开始年份（所有文献）',marker='.',markersize=8)
ax.plot(KeyStartYear,np.cumsum(np.array(freq3))/5866212.0, label='开始年份（主要负责文献）',marker='.',markersize=8)
ax.plot(CommonEndYear,np.cumsum(np.array(freq2))/5866212.0,linestyle='dashed', label='暂停年份（所有文献）',marker='.',markersize=8)
ax.plot(KeyEndYear,np.cumsum(np.array(freq4))/5866212.0,  label='暂停年份（主要负责文献）',marker='.',markersize=8)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=0))
ax.set(xlabel='作者职业生涯开始/暂停年份', ylabel='累计比例',
       title='作者职业生涯开始/暂停年份累计比例')
# plt.ticklabel_format(style='plain')
# 5866212
plt.text(-0.03,1.05,'总人数:586万',
horizontalalignment='center',verticalalignment='center',
transform=ax.transAxes)
ax.set_xlim(1980, 2022)
ax.set_yticks(np.array([ i for i in range(0, 11, 1)])/10)
# ax.set_xticks([2000,2005,2,3])
# ax.set_ylim(-0.001, 0.41)
# ax.set_yscale('log')
ax.legend(fontsize=12,framealpha=0.7,facecolor='white',frameon=True)
fig.savefig("作者职业生涯开始或暂停年份.png",dpi=300)
plt.show()
```

```python
# 开始或终止年份
CommonSpanYear , freq1 = zip(*sorted(CommonSpanYearCounter.items()))
KeySpanYear , freq2 = zip(*sorted(KeySpanYearCounter.items()))

fig, ax = plt.subplots()
ax.plot(np.array(CommonSpanYear)+1,np.cumsum(np.array(freq1))/5866212.0,  linestyle='dashed', label='职业生涯持续年份（所有文献）',marker='.',markersize=8)
ax.plot(np.array(KeySpanYear)+1,np.cumsum(np.array(freq2))/5866212.0, label='职业生涯持续年份（主要负责文献）',marker='.',markersize=8)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0, decimals=0))
ax.set(xlabel='作者职业生涯持续年份', ylabel='累计比例',
       title='作者职业生涯持续年份累计比例')
# plt.ticklabel_format(style='plain')

plt.text(-0.03,1.05,'总人数:586万',
horizontalalignment='center',verticalalignment='center',
transform=ax.transAxes)
ax.set_xlim(0.5, 32)
ax.set_xticks([ i for i in range(2, 33, 2) ])
ax.set_yticks(np.array([ i for i in range(0, 11, 1)])/10)

# ax.set_ylim(0, 0.0002)
# ax.set_yscale('log')
ax.legend(fontsize=12,framealpha=0.7,facecolor='white',frameon=True)
fig.savefig("作者职业生涯持续年份.png",dpi=300)
plt.show()
```
